### **# Load Dataset**

```python
# Load Dataset
import pandas as pd

df = pd.read_csv("Customer Personality Analysis.csv")

print(df.head())
```

### **# Check Missing Values**

```python
# Check Missing Values
df.isnull().sum()
```

```python
df['Income'].fillna(df['Income'].median(), inplace=True)
```

```python
df.isnull().sum()
```

### **# Duplicates Remove**

```python
df.drop_duplicates()
```

```python
print(df.duplicated().sum())
```

### **Data Cleaning**

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
```

```python
current_year = 2026

df['Age'] = current_year - df['Year_Birth']
```

```python
df[['Year_Birth','Age']].head()
```

```python
# Create Total Spending
df['Total_Spending'] = (
    df['MntWines'] +
    df['MntFruits'] +
    df['MntMeatProducts'] +
    df['MntFishProducts'] +
    df['MntSweetProducts'] +
    df['MntGoldProds']
)
```

```python
df[['Total_Spending']].head()
```

```python
# Create Total Children
df['Children'] = df['Kidhome'] + df['Teenhome']
```

```python
# Campaign Analysis
campaigns = [
    'AcceptedCmp1',
    'AcceptedCmp2',
    'AcceptedCmp3',
    'AcceptedCmp4',
    'AcceptedCmp5'
]

df[campaigns].sum()
```

```python
# Customer Segmentation
df.sort_values(
    by='Total_Spending',
    ascending=False
)[['ID','Income','Total_Spending']].head(10)
```

```python
# age group spends the most
df.groupby('Age')['Total_Spending'].mean()
```

```python
# income affect spending
df[['Income','Total_Spending']].corr()
```

```python
# Top Product Category
products = [
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds'
]
df[products].sum()
```

```python
# campaign got the highest acceptance
df[campaigns].sum()
```

```python
# customers with children spend less
df.groupby('Children')['Total_Spending'].mean()
```

```python
# Income Distribution
plt.figure(figsize=(7,5))
sns.histplot(df['Income'], bins=30, kde=True)
plt.title('Income Distribution')
plt.show()
```

```python
# Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=25)
plt.title('Age Distribution')
plt.show()
```

```python
# top product category
plt.figure(figsize=(6,5))
df[products].sum().sort_values().plot(kind='bar')
plt.title('Product Sales')
plt.show()
```

```python
# Marital Status Count
plt.figure(figsize=(6,5))
sns.countplot(x='Marital_Status', data=df)
plt.title('Marital Status Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Income vs Spending
plt.figure(figsize=(6,5))
plt.title('Income vs Spending')
sns.scatterplot(
    x='Income',
    y='Total_Spending',
    data=df
)
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(7,5))
plt.title('Customer Behavior Correlation Heatmap')

sns.heatmap(
    df.corr(numeric_only=True),
    cmap='coolwarm'
)

plt.show()
```

```python
# Campaign Analysis
plt.figure(figsize=(6,5))
df[campaigns].sum().plot(kind='bar')
plt.title('Campaign Acceptance')
plt.show()
```
